## 1. Определить корни

```python
from sympy import *

a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol("x")

f_x = a * (x**3) + b * (x**2) + c * (x**1) + d
```

```python
f_x
```

```python
import numpy as np

coeff = [a, b, c, d]
print("Корень 1: ", np.roots(coeff)[0])
print("Корень 2: ", np.roots(coeff)[1])
print("Корень 3: ", np.roots(coeff)[2])
```

## 4. Построить график

```python
from sympy.plotting import plot

g1 = plot(f_x,(x, -7, 1), line_color = 'red', title = 'график функции')
```

```python
derivative = diff(f_x)
derivative
```

```python
g2 = plot(derivative,(x, -7, 1), line_color = 'green', title = 'график производной функции')
```

```python
g3 = plot(f_x, derivative,(x, -7, 1), title = 'Совмещенный график', legend = True, xlabel = 'X', ylabel = 'Y', show = False)

g3[0].line_color = 'red'
g3[1].line_color = 'green'

g3.show()
```

```python
print("Вершина: ", solve(derivative, x)[0].n(3))
```

```python
derivative_2 = diff(derivative)
derivative_2
```

```python
g4 = plot(derivative_2,(x, -7, 1), line_color = 'black', title = 'график второй производной функции')
```

```python
g5 = plot(f_x, derivative, derivative_2, (x, -7, 1), title = 'Совмещенный график', legend = True, xlabel = 'X', ylabel = 'Y', show = False)

g5[0].line_color = 'red'
g5[1].line_color = 'green'
g5[2].line_color = 'black'

g5.show()
```

```python
solve(derivative_2 > 0, x)
```

```python
solve(derivative_2 < 0, x)
```

```python
solve(derivative > 0, x)
```

```python
solve(derivative < 0, x)
```

```python
solve(f_x > 0, x)
```
